```python
from lxml import etree
from io import StringIO
import requests
import pandas as pd
from bs4 import BeautifulSoup
df = pd.DataFrame(columns=['date','Rank','Title', 'Artist','Weeks on chart'])
ind = 0
listofdata = []
for year in range(1955,1958):
    print year
    for month in range(1,13):
        for date in range(1,28,7):
            first = True
            string = str(year)+"-"+str(month).zfill(2)+"-"+str(date).zfill(2)
            page = requests.get('https://www.billboard.com/charts/hot-100/'+string)
            soup = BeautifulSoup(page.content,"lxml")
            title = soup.find('div',class_="chart-number-one__title").text.strip()
            artist = soup.find('div',class_="chart-number-one__artist").text.strip()
            weeks = soup.find('div',class_="chart-number-one__weeks-on-chart").text.strip()
            #print string,title.text,artist.text,weeks.text
            df.loc[ind] = [string,"1",title,artist,weeks]
            ind +=1
            mydivs = soup.find_all("div",class_="chart-list-item")
            for div in mydivs:
                rank = div['data-rank'].encode('utf-8')
                title = div['data-title'].encode('utf-8')
                artist = div['data-artist'].encode('utf-8')
                week = div.find_next('div',class_="chart-list-item__weeks-on-chart")
                if not week:
                    week=  None
                else:
                    week = week.text.strip().encode('utf-8')
                df.loc[ind] = [string,rank,title,artist,week]
                ind +=1
df.to_csv('weekly_billboard_1955_1958.csv',index=False)
```
