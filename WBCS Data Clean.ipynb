```python
def convert(number):
    return round(number)

new_df['Rank'] = new_df['Rank'].apply(convert)
new_df.head()
```

```python
import re
def split_it(song):
    string = re.split(r'[:]', song)
    return string[0]

new_df['Artist'] = new_df['Song'].apply(split_it)
```

```python
def split_it(song):
    string = re.split(r'[:]', song)
    return string[1]

new_df['Song1'] = new_df['Song'].apply(split_it)
new_df.head()
```

```python
def split_it(song):
    string = re.split(r'[:]', song)
    return string[2]

new_df['Song2'] = new_df['Song'].apply(split_it)
new_df.head()
```

```python
def split_it(song):
    string = re.split(r'[#]', song)
    return string[0]

new_df['Song2'] = new_df['Song2'].apply(split_it)
new_df.head()
```

```python
def space(string):
    length = len(string)
    return string[:length-1]

new_df['Artist'] = new_df['Artist'].apply(space)
new_df['Song1'] = new_df['Song1'].apply(space)
new_df.head()
```

```python
new_df = new_df.drop('Song',axis=1)
```

```python
def split_it(year):
    string = re.findall(r'(\d{4})', year)
    return string[0]
     
new_df['Year'] = new_df['Year'].apply(split_it)
new_df.head()
```

```python
new_df
```

```python
def rspace(string):
    l = len(string)
    return string[1:l]

new_df['Song1']=new_df['Song1'].apply(rspace)
new_df['Song2']=new_df['Song2'].apply(rspace)
```
