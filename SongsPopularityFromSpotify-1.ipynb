```python
import pandas as pd
import numpy as np
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pprint

client_credentials_manager = SpotifyClientCredentials('9642c853b7ee433c96833d95052ae7f3', '6574d7e8082e461bab72f83978d7f2f3')
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
pp = pprint.PrettyPrinter(indent=1)

songs = pd.read_csv("top100billboard.csv",sep=",")
#num = np.arange(200)
#x = songs.iloc[num]

popularity = []
for index, row in songs.iterrows():
    q = 'track:"'+row['Title'] + '" artist:"' + row['Artist']+'" year:1940-'+str(row['year'])
    results = sp.search(q=q, limit=20, type='track', market='US')
    maxm = -1
    chosenRow = None
    for i, t in enumerate(results['tracks']['items']):
        if(t['popularity'] > maxm):
            maxm = t['popularity']
            chosenRow = t
    
    if chosenRow is not None:
        print(chosenRow['name'], row['year'], chosenRow['popularity'])
        popularity.append(chosenRow['popularity'])
    else:
        popularity.append(-1)

songs['Spotify_Popularity'] = pd.Series(popularity).astype(int)
songs.to_csv("songsWithPopularity.csv", index=False)
    
        
```
