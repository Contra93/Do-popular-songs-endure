```python
#!/usr/bin/env python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.plotly as py
import plotly.graph_objs as go
from matplotlib import pylab
from scipy import stats

songs = pd.read_csv("weeklySongsWithPopularity.csv",sep=",")
songs = songs[songs["Spotify_Popularity"] > 0]
songs['date'] = pd.to_datetime(songs['date'],format="%Y-%m-%d")
songs['year'] = songs['date'].dt.year

plt.rcParams['figure.figsize'] = [50.0, 40.0]
fig=plt.figure(1)
fig.subplots_adjust(hspace=.5)
fig.add_subplot(521)

for i in range(1,6):
    ranksongs = songs[songs['Rank']==i]
    trainX = ranksongs["year"]
    trainY = ranksongs["Spotify_Popularity"]
    plt.subplot(7, 2, i)
    plt.plot(trainX, trainY, 'o')
    #plt.scatter(trainX, trainY, alpha=0.5,s=10)
    plt.title('Year vs popularity for Rank '+str(i)+' Songs')
    plt.xticks(np.arange(1958, 2018+1, 5.0))
    plt.yticks(np.arange(0, 100, 10.0))
    plt.xlabel('Year')
    plt.ylabel('Popularity')
    plt.show()
```

```python
plt.rcParams['figure.figsize'] = [50.0, 40.0]
fig=plt.figure(1)
fig.subplots_adjust(hspace=.5)
fig.add_subplot(521)

for i in range(100,94,-1):
    ranksongs = songs[songs['Rank']==i]
    trainX = ranksongs["year"]
    trainY = ranksongs["Spotify_Popularity"]
    plt.subplot(7, 2, 101-i)
    plt.plot(trainX, trainY, 'o')
    #plt.scatter(trainX, trainY, alpha=0.5,s=10)
    plt.title('Year vs popularity for Rank '+str(i)+' Songs')
    plt.xticks(np.arange(1958, 2018+1, 5.0))
    plt.yticks(np.arange(0, 100, 10.0))
    plt.xlabel('Year')
    plt.ylabel('Popularity')
    plt.show()
```

```python
#mean of spotify popularity year wise
plt.rcParams['figure.figsize'] = [10.0, 5.0]
meanPop = songs.groupby('year').mean().reset_index()
plt.plot(meanPop['year'], meanPop['Spotify_Popularity'],'-o')
plt.xlabel('Year')
plt.ylabel('Average Current Popularity')
```
