```python
import pandas as pd
import numpy as np
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pprint
```

```python
songs = pd.read_csv("weekly_data_with_year_1955_2017.csv",sep=",")

visited = {}
uniqueArtists = []
for index, row in songs.iterrows():
    if row['Artist'] not in visited.keys():
        visited[row['Artist']] = 1
        uniqueArtists.append(row['Artist'])

len(uniqueArtists)
```

```python
client_credentials_manager = SpotifyClientCredentials('9642c853b7ee433c96833d95052ae7f3', '6574d7e8082e461bab72f83978d7f2f3')
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
pp = pprint.PrettyPrinter(indent=1)

popularityPerArtist = {}
for row in uniqueArtists:
    q = 'artist:"' + row+'" '
    results = sp.search(q=q, limit=20, type='artist', market='US')
   
    maxm = -1
    chosenRow = None
    
    for i, t in enumerate(results['artists']['items']):
        if(t['popularity'] > maxm):
            maxm = t['popularity']
            chosenRow = t
    
    if chosenRow is not None:
        popularityPerArtist[row] = maxm
    else:
        popularityPerArtist[row] = -1
    
    print(row, popularityPerArtist[row])

len(popularityPerArtist)
```

```python
songs = pd.read_csv("weeklySongsWithPopularity.csv",sep=",")
```

```python
artistPopularityList = []
for index, row in songs.iterrows():
    artistPopularityList.append(popularityPerArtist[row['Artist']])

songs['Artist_Popularity'] = pd.Series(artistPopularityList).astype(int)
```

```python
songs.tail()
```

```python
songs.to_csv("weeklySongsWithSongAndArtistPopularity.csv", index=False)
```
